## 1- Import Libraries

```python
import cv2 as cv
import numpy as np
```

## 2- Load our first image

```python
img = cv.imread('input.jpg')

cv.imshow('It is a cow!', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

## 3- Saving Image

```python
type(img)
```

```python
print(img.shape)
```

### Printing each dimension

```python
print('Height of Image: ', int(img.shape[0]), 'pixels')
print('Width of Image: ', int(img.shape[1]), 'pixels')
```

### Saving

```python
cv.imwrite('output.jpg', img)
cv.imwrite('output.png', img)
```

## 4- Displaying in matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

## 5- Change Color Channels (BGR -> RGB)

### 1- using numpy

```python
# plt.imshow(img[:,:,::-1])
plt.imshow(img[...,::-1])
```

### 2- Using OpenCV

```python
rgb_image = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

## 6- Change and manipulate pixels

- Method 1

```python
img = cv.imread('input.jpg')
img[200:300, 300:500] = 255
plt.imshow(img[...,::-1])
```

- Method 2

```python
img[200:300,300:500,:] = (0, 0, 255)
plt.imshow(img[...,::-1])
```
