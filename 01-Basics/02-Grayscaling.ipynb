## 1- Converting Color image to Grayscale

- Method 1

```python
import cv2 as cv

img = cv.imread('./images/input.jpg')
cv.imshow('Original', img)
cv.waitKey()

gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

cv.imshow('Grayscale', gray_img)
cv.waitKey()
cv.destroyAllWindows()
```

- Method 2

```python
img = cv.imread('./images/input.jpg', 0)
cv.imshow('Grayscale', img)
cv.waitKey()
cv.destroyAllWindows()

print(img.shape)
```

## 3- Check Dimensions

```python
img = cv.imread('./images/input.jpg')

B, G, R = img[10, 50]

print(B, G, R)
print("shape:", img.shape)
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

print(gray_img[10, 50])
print("shape:", gray_img.shape)
```
