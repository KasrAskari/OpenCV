## 1- Import Libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## 2- Loding Image

```python
img = cv.imread('images/input.jpg')
plt.imshow(img[...,::-1])
```

## 3- Cropping

- Method 1

```python
height, width = img.shape[:2]

cropped = img[250:400, 250:550]

plt.imshow(cropped[...,::-1])
```

```python
cv.imshow("original", img)
cv.imshow("cropped", cropped)

cv.waitKey()
cv.destroyAllWindows()
```

- Method 2

```python
height, width = img.shape[:2]

cropped = img[height//3:3*height//4, width//3:3*width//4]

plt.imshow(cropped[...,::-1])
```
