## 1- Importing Libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## 2- Loading Image

```python
img = cv.imread("IMG_4720.jpg")
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.show()
```

## 3- Finding Face Coordinates

```python
height, width = img_rgb.shape[:2]

face = img_rgb[500:2400, 600:2000]

plt.imshow(face)
plt.title("Face Image")
plt.show()
```

## 4- Save New Image

```python
face_rgb = cv.cvtColor(face, cv.COLOR_BGR2RGB)

face_path = "face_img.jpg"
cv.imwrite(face_path, face_rgb)
```
