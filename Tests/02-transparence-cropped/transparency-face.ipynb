## 1- Importing Libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## 2- Loading Image

```python
img = cv.imread("IMG_4720.jpg")
face = img[450:2500, 150:2500]

circle = np.zeros((face.shape[0], face.shape[1], 3), np.uint8)
cv.circle(circle, (face.shape[1]//2, face.shape[0]//2), min(face.shape[0], face.shape[1])//2, (255, 255, 255), -1)

output_img = cv.bitwise_and(face, circle)
plt.imshow(output_img[...,::-1])
```

## 3- Saving "output_img"

```python
png_img = np.zeros((face.shape[0], face.shape[1], 4), np.uint8)

png_img[:,:,0:3] = output_img

circle_gray = cv.cvtColor(circle, cv.COLOR_BGR2GRAY)
png_img[:,:,3] = circle_gray


cv.imwrite("face.png", png_img)
plt.imshow(cv.cvtColor(png_img, cv.COLOR_BGR2RGB))
plt.show()
```
