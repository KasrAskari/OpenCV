## 1- Importing Libaries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## 2- Drawing Rectangle & Ploylines

```python
img = np.zeros((500, 500, 3), dtype=np.uint8)

# Reactangle
start_point = (150, 300)  
end_point = (350, 450)    
color = (255, 255, 0)         
thickness = 3
cv.rectangle(img, start_point, end_point, color, thickness)

# Polynomial
triangle_points = np.array([[100, 300], [250, 150], [400, 300]], np.int32)
triangle_points = triangle_points.reshape((-1, 1, 2))
color = (250, 0, 0)  
thickness = 3  
cv.polylines(img, [triangle_points], isClosed=True, color=color, thickness=thickness)

image_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.figure(figsize=(6, 6))
plt.imshow(image_rgb)
plt.show()
```
