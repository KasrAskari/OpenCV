## 1- Importing Libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## 2- Loading Image

```python
img_original = cv.imread('Weeki_Wachee_spring_10079u.jpg')
```

## 3- Linear Adjustment

```python
output_linear = cv.convertScaleAbs(img_original, alpha=1.5, beta=50)

# plt.imshow(output_linear[...,::-1])
```

## 4- Gamma Correction

```python
gamma = 0.5

lookUpTable = np.empty((1, 256), np.uint8)

for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma)*255.0, 0, 255)
    
    output_gamma_corrected = cv.LUT(img_original, lookUpTable)
    
# plt.imshow(output_gamma_corrected[...,::-1])
```

## 5- Linear vs Gamma

```python
plt.figure(figsize=[10,6])

plt.subplot(231)
plt.imshow(img_original[...,::-1])
plt.title("Original")

plt.subplot(232)
plt.imshow(output_linear[...,::-1])
plt.title("Linear")

plt.subplot(233)
plt.imshow(output_gamma_corrected[...,::-1])
plt.title("Gamma")
```
