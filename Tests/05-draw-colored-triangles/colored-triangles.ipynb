## 1- Importing Libraries

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

## 2- Drawing Triangles

```python
color = (0,0,255)

def draw_triangle(event, x, y, flags, param):
    global color
    if event == cv.EVENT_LBUTTONDOWN:
        point1 = (x, y - 20)
        point2 = (x - 20, y + 20)
        point3 = (x + 20, y + 20)
        triangle_cnt = np.array([point1, point2, point3])

        cv.drawContours(img, [triangle_cnt], 0, color, -1)
        
img = np.zeros((512,512,3), np.uint8)
cv.namedWindow('image')
cv.setMouseCallback('image', draw_triangle)

while True:
    cv.imshow('image', img)
    k = cv.waitKey(1) & 0xFF
    
    if k == ord('b'):
        color = (255,0,0)
    elif k == ord ('g'):
        color = (0,255,0)
    elif k == ord('r'):
        color = (0,0,255)
    elif k == ord('y'):
        color = (0,255,255)
    elif k == ord('p'):
        color = (220,195,255)
    elif k == ord('w'):
        color = (255,255,255)
    elif k == 27:
        break
    
cv.destroyAllWindows()
plt.imshow(img[...,::-1])
```
